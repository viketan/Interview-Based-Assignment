```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
```

```python
bp_before = df[' Blood Pressure Before (mmHg)']
bp_after = df[' Blood Pressure After (mmHg)']
```

```python
# Perform Shapiro-Wilk test
_, p_before = stats.shapiro(bp_before)
_, p_after = stats.shapiro(bp_after)
```

```python
# Set significance level
alpha = 0.05

# Print the p-values and interpret the results
print("Shapiro-Wilk Test Results:")
print(f"Blood Pressure Before: p-value = {p_before}")
print(f"Blood Pressure After: p-value = {p_after}")

if p_before > alpha and p_after > alpha:
    print("The change in blood pressure follows a normal distribution.")
else:
    print("The change in blood pressure does not follow a normal distribution.")
```

```python
# Create the plot
plt.figure(figsize=(8, 6))
plt.hist(bp_before, bins=10, edgecolor='black')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.title('Distribution of Blood Pressure Before')
```

```python
# Create the plot
plt.figure(figsize=(8, 6))
plt.hist(bp_after, bins=10, edgecolor='black')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.title('Distribution of Blood Pressure After')
```

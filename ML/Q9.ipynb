```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
drop_columns = [col for col in data.columns if data[col].isna().sum()/len(data[col])*100>10]
```

```python
df = data.drop(drop_columns,axis=1)
```

```python
cat = [col for col in df.columns if df[col].dtype=='object']
num = [col for col in df.columns if df[col].dtype!='object']
```

```python
for col in cat:
    df[col] = df[col].fillna(df[col].mode().iloc[0])
```

```python
for col in num:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isna().sum().sum()
```

```python
df.head()
```

```python
drop = []
for col in cat:
    count = df[col].value_counts()
    if len(count)>100:
        drop.append(col)
    print(f"{col}:{count}")
```

```python
drop
```

```python
df.drop(drop,axis=1,inplace=True)
```

```python
df.head()
```

```python
# Split the DataFrame into two subsets based on the HasDetections
df_yes = df[df['HasDetections'] == 1]
df_no = df[df['HasDetections'] == 0]
```

```python
# Randomly sample an equal number of rows from each subset
sample_size = int(0.05 * len(df))
random_sample_yes = df_yes.sample(n=sample_size, replace=False)
random_sample_no = df_no.sample(n=sample_size, replace=False)
```

```python
# Combine the sampled subsets into a single DataFrame
random_sample = pd.concat([random_sample_yes, random_sample_no])
```

```python
X = random_sample.drop('HasDetections',axis=1)
y = random_sample['HasDetections']
```

```python
random_sample.shape
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, log_loss
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
confusion_mat = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
logloss = log_loss(y_test, model.predict_proba(X_test))

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion_mat)
print("Classification Report:\n", classification_rep)
print("ROC AUC Score:", roc_auc)
print("Log Loss:", logloss)
```
